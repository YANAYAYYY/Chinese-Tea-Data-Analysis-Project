```python
#提取PDF中所有页的表格（Extract a table of all pages in the PDF）

f = pdfplumber.open(r'D:\就业思路\数据分析学习\作品集\茶叶数据分析作品集\资料收集\2022年中国茶叶产销形势报告2.pdf')

all_tables = []
tables = []

for i in range(len(f.pages)):
    page_tables = f.pages[i].extract_tables()
    
    for table in page_tables:
        df = pd.DataFrame(table[1:], columns = table[0])
        all_tables.append(df)
        tables.append(table)
        
print(all_tables)
```

```python
print(tables)
```

```python
#——————第一个表格:2022年度全国茶园总面积(单位：万亩)---------------
table1 = all_tables[0]
table1
```

```python
#修改文字识别的错误，如"湖北"被识别成了“湖才七”。（Corrected an error in character recognition, for example, "湖北" was identified as "湖才七".）

table1.iloc[7, 0] = '湖北'   #[行号，列号]

#查看第一个表格（check the first table）
print("2022年度全国茶园总面积(单位：万亩)")
print(table1)
```

```python
#将table1保存到excel（Save the table1 to excel）

table1.to_excel('1 The total area of tea plantations in China in 2022.xlsx', sheet_name='sheet 1', index=False)
```

```python
#——————第二个表格:2022年度全国干毛茶总产量（单位：吨）---------------
#（The total productiont of dried rough tea in China in 2022 (unit: tons)）

table2 = all_tables[1]
table2
```

```python
#问题：第二个表格分别在两页上，所以程序将他们断成了两个表格。合并两个表格（Merge the two tables）

#查看剩下的表格

table2_column_names = table2.columns.tolist()

table2_nextpage = pd.DataFrame(tables[2],columns = table2_column_names)

table2_nextpage
```

```python
#合并两个表格
table2_all = pd.concat([table2, table2_nextpage], ignore_index=True)

#修改错误
table2_all.iloc[7, 0] = '湖北'  #[行号，列号]

print("2022年度全国干毛茶总产量(单位：吨)")
print(table2_all)
```
