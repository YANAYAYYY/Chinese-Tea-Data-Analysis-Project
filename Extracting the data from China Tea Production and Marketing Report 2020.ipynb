```python
import pdfplumber

from tabula import read_pdf

import xlwings as xw

import pandas as pd
```

```python
#提取PDF中所有页的表格（Extract a table of all pages in the PDF）

f = pdfplumber.open(r'D:\就业思路\数据分析学习\作品集\茶叶数据分析作品集\资料收集\2020年中国茶叶产销形势报告2.pdf')

all_tables = []
tables = []
column_names = []
row_names = []

for i in range(len(f.pages)):
    page_tables = f.pages[i].extract_tables()
    
    for table in page_tables:
        df = pd.DataFrame(table[1:], columns = table[0])
        #提取column
        column_name = df.columns.tolist()
        #提取row
        row_name = df.iloc[:,0].tolist()
        column_names.append(column_name)
        row_names.append(row_name)
        all_tables.append(df)
        tables.append(table)
        
print(all_tables[:3])
```

```python
f2 = pdfplumber.open(r'D:\就业思路\数据分析学习\作品集\茶叶数据分析作品集\资料收集\2020年中国茶叶产销形势报告8.pdf')

all_data = []
data = []

for i in range(len(f2.pages)):
    page_tables = f2.pages[i].extract_tables()
    
    for table in page_tables:
        df_data = pd.DataFrame(table[1:], columns = table[0])
        all_data.append(df_data)
        data.append(table)
        
print(all_data[:3])
```
